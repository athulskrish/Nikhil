```python
import random
import string
```

```python
alphabet= string.ascii_uppercase
print(alphabet)
```

```python
#random nodes
nodes=[]

for i in range(2,random.randint(5,10)) :
    nodes.append(random.choice(alphabet))
    print(nodes)
```

```python
#random adjaceny list

adj_lst={}
for i in nodes:
  relation=[]
  for j in range(0,random.randint(1,len(nodes))):

    relative=random.choice(nodes)
    if relative not in i :
     relation.append(relative)
  adj_lst[i]=relation
print(adj_lst)

```

```python
#drawing the random graph

random_graph=nx.DiGraph(adj_lst)
nx.draw(random_graph, with_labels=True,node_color ='lightblue', font_size=10, node_size=1000)
plt.show()
```

```python
#random adjaceny list with weights

adj_lst2={}

for i in nodes:
  relation=[]


  relative=random.choice(nodes)
  weight=random.randint(1,10)
  while relative== i:
    relative=random.choice(nodes)

  node_weights=(relative, weight)
#     node_weights.add(relative)
#     node_weights.add(weight)
  relation.append(node_weights)

    #print(relative)
  adj_lst2[i]=relation

print(adj_lst2)

```

```python
#drawing the random graph
random_graph=nx.Graph()

for node, edges in adj_lst2.items():
  for edge in edges:
    random_graph.add_edge(node, edge[0],weight=edge[1])

labels=nx.get_edge_attributes(random_graph, 'weight')
pos=nx.spring_layout(random_graph)
nx.draw(random_graph,pos, with_labels=True,node_color ='lightblue', font_size=10, node_size=1000)


nx.draw_networkx_edge_labels(random_graph,pos,edge_labels=labels)


plt.show()
```

```python
#shortest path
start_node = random.choice(list(random_graph.nodes()))
end_node = random.choice(list(random_graph.nodes()))

try:
  shortest_path = nx.shortest_path(random_graph, source=start_node, target=end_node)
  print(f"Shortest path from {start_node} to {end_node}: {shortest_path}")
except nx.NetworkXNoPath:
  print(f"There is no path between {start_node} and {end_node}")
```
